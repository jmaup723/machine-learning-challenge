```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#list(df.columns) 
df.dtypes
```

```python
X = df.drop("koi_disposition", axis=1)
```

```python
y = df["koi_disposition"]
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)

X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
X_minmax = MinMaxScaler().fit(X_train)

X_train_minmax = X_minmax.transform(X_train)
X_test_minmax = X_minmax.transform(X_test)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python

clf = clf.fit(X_train_minmax, y_train)
```

```python
print(f"Training Data Score: {clf.score(X_train_minmax, y_train)}")
print(f"Testing Data Score: {clf.score(X_test_minmax, y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV
dt_model = tree.DecisionTreeClassifier()
```

```python
dt_param_grid = {'max_depth':[3,9,11,15], 'min_samples_leaf':[5,10,35,60,180]}
```

```python
dt_grid = GridSearchCV(dt_model, dt_param_grid, verbose=3)
dt_grid.fit(X_train_minmax, y_train)
print("fin")
```

```python
dt_predictions= dt_grid.predict(X_test_minmax)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, dt_predictions))
```

```python
print(dt_grid.best_params_)
print(dt_grid.best_score_)
```
